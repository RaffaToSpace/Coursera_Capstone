Introducing the function to collect information about venues in neighbourhoods

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Analyse neighbourhoods
Using one hot encoding to organise information about venue categories.

```python
# one hot encoding
trno_c_onehot = pd.get_dummies(trno_c_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
trno_c_onehot['Neighborhood'] = trno_c_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [trno_c_onehot.columns[-1]] + list(trno_c_onehot.columns[:-1])
trno_c_onehot = trno_c_onehot[fixed_columns]

trno_c_onehot.head()
```

```python
trno_c_onehot.shape
```

### Grouping rows by neighborhood and taking the mean of the frequency of occurrence of each category

```python
trno_c_grouped = trno_c_onehot.groupby('Neighborhood').mean().reset_index()
trno_c_grouped.head()
```

Printing the top 5 venues of each neighbourhood.

```python
num_top_venues = 5

for hood in trno_c_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = trno_c_grouped[trno_c_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

We are now going to put the information in a dataframe.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
trno_c_venues_sorted = pd.DataFrame(columns=columns)
trno_c_venues_sorted['Neighborhood'] = trno_c_grouped['Neighborhood']

for ind in np.arange(trno_c_grouped.shape[0]):
    trno_c_venues_sorted.iloc[ind, 1:] = return_most_common_venues(trno_c_grouped.iloc[ind, :], num_top_venues)

trno_c_venues_sorted.head()
```

### Clustering neighbourhoods
Using k means to cluster the neighbourhoods.

```python
kclusters = 5

trno_c_grouped_clustering = trno_c_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(trno_c_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:40]
```

```python
trno_centre
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
trno_c_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

tno_c_merged = trno_centre

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
tno_c_merged = tno_c_merged.join(trno_c_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

tno_c_merged.head()
```

Let's now make a clusters map.

```python
# create map
map_trno_c_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(tno_c_merged['Latitude'], tno_c_merged['Longitude'], tno_c_merged['Neighbourhood'], tno_c_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_trno_c_clusters)
       
map_trno_c_clusters
```

### Understanding the clustering outcomes
From the cluster tables it is very clear how the majority of neighbourhoods are clustered together in cluster 0 as they feature Cafes, Coffee shops and other similar public venues as most popular venues.

The other clusters are more diverse and made of 1 or 2 elements at most. We can thus say that neighbourhoods in central Toronto are all fairly similar to each other, with some exceptions where parks, transportation infrastructure, restaurants or other services are more concentratited.
