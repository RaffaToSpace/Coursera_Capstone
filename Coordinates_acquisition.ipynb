Two columns for the coordinates are added, and the values are filled using the "get_coordinates" function previously defined.

```python
columns_list = ["Code","Borough","Neighbourhood","Latitude","Longitude"]
toronto_table = toronto_table.reindex(columns = columns_list)
```
